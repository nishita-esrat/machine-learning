```python
import seaborn as sns
import matplotlib.pyplot as pl
```

### lmplot() <br>

##### regression plot is used in machine learning so we will not fully discuss what, how and where it is used but we will see how 
##### it is created and what are the values we can put inside it

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips)
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,hue = 'sex')
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,hue = 'sex',palette='coolwarm')
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,hue = 'sex',palette='coolwarm',markers=['o','v'],
           scatter_kws={'s':50})
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,hue = 'sex',palette='coolwarm',markers=['o','v'])
```
