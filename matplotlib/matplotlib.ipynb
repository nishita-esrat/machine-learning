```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Matplotlib Basics

```python
x = np.linspace(0,5,11)
x
y = x **2 
y
```

```python
plt.plot(x,y)
plt.title("plot y = X^2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

### Subplots in Matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```

### Object Oriented way

#### Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

#### figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).

```python
fig = plt.figure(figsize=(20,8),dpi = 70)
axis1 = fig.add_axes([0.1,0.1,.4,.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,.3,.3])
axis2.plot(x,y)
```
